```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("1.csv")
print(data)
```

```python
x=data.drop(columns=["medv"])
y=data["medv"]
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.33,random_state = 42)
```

```python
print(x_train.shape)
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense
```

```python

#1. model initialization

model = Sequential() # Empty model or a plain canvas without any HL's or inputs
model.add(Input(shape=(13,))) # Input layer
model.add(Dense(units = 3, activation = "relu"))
model.add(Dense(units =1)) # Dense layer or a fully connected layer, means  all units in the prev. layer are connected to every unit of the current layer

```

```python
# model compilation
#adam ,rmsprop , Adadelta , Adagrad
model.compile(optimizer = "rmsprop", loss = "mean_squared_error")
```

```python
# model ttraining/ fit
model.fit(x= x_train,y = y_train, epochs = 150, validation_data=(x_test , y_test))  # epochs are the number of iterations.
```

```python
model.history.history.keys()
```

```python
train_loss = model.history.history["loss"]
val_loss = model.history.history["val_loss"]

plt.plot(train_loss)
plt.plot(val_loss)
plt.xlabel("Epochs")
plt.ylabel("MSE")
# plt.ylim((0,250))
plt.legend(["train","test"])
plt.grid()
plt.show()
```
